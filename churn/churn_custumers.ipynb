Посмотрим на соотношение классов в столбце "Пол" и "Страна"

```python
#функия красиво вывести количество строк
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}% \n кол:{v:d}'.format(p=pct,v=val)
    return my_autopct
```

## Исследование задачи

Чтож начнем,по видиму этот проект задуман так,что мы сначала делаем при дисбалансе классов,а потом боремся с проблемой.Так и поступим.Сначала взглянем на сам целевой диссбаланс

Дерево решений 

Начнем с классики - дерева решений,попробуем разную глубину и построим графики зависимости глубины от f1-меры,найдем лучшие гипперпараметры 

```python
#поиск лучшего 
best_tree = None
best_depth_tree = 0
best_f1_tree = 0
best_auc_roc_tree = 0
# для графиков 
depth_tree = []
f1_train_tree = []
f1_valid_tree = []
auc_roc_tree_list = []
auc_roc_tree_list_features =[]
#создаем 
for depth in range(1, 21, 1):
    model = DecisionTreeClassifier(random_state=12345, max_depth=depth)
    model.fit(features_train, target_train) 

    # получаем  f1-меру для обучающей и валидационной 
    pred_train_tree = pd.Series( model.predict(features_train))
    f1_train = f1_score(target_train, pred_train_tree)  

    pred_valid_tree = pd.Series(model.predict(features_valid))
    f1_valid = f1_score(target_valid, pred_valid_tree)   

    # добовляем 
    depth_tree.append(depth)
    f1_train_tree.append(f1_train)
    f1_valid_tree.append(f1_valid)
    
    #ищем auc_roc
    prob_valid = model.predict_proba(features_valid)
    prob_one_valid = prob_valid[:, 1]
    auc_roc_tree = roc_auc_score(target_valid,prob_one_valid)
    auc_roc_tree_list.append(auc_roc_tree)

    prob_features_train = model.predict_proba(features_train)
    prob_one= prob_features_train[:, 1]
    auc_roc_tree_f = roc_auc_score(target_train,prob_one)
    auc_roc_tree_list_features.append(auc_roc_tree_f)
    #ищем лучшую
    if f1_valid > best_f1_tree:
      best_tree = model
      best_depth_tree = depth
      best_f1_tree = f1_valid
      best_auc_roc_tree = auc_roc_tree
print('лучшая модель "дерево решений":',best_tree)
print('Глубина дерева "дерево решений":',best_depth_tree)
print('F1-мера "дерево решений":',best_f1_tree)
print('AUC-ROC лучшей модели "дерево решений":',best_auc_roc_tree)     
```

```python
plt.figure(figsize=(12, 4))
plt.grid(True)
plt.xticks(depth_tree)
plt.plot(depth_tree, f1_train_tree,'-bo')
plt.plot(depth_tree,f1_valid_tree,'-ro')
plt.legend(['Обучающая выборка', 'Валидационная выборка'])
plt.xlabel('Глубина дерева')
plt.ylabel('F1-мера')
plt.title('Зависимость F1-меры модели "дерево решений"  от глубины дерева')
plt.show()
```

```python
x = depth_tree
y = auc_roc_tree_list
y1 = auc_roc_tree_list_features
plt.figure(figsize=(12, 4))
plt.grid(True)
plt.xticks(x)
plt.plot(x, y,'-bo')
plt.plot(x, y1,'-ro')
plt.xlabel('depth')
plt.ylabel('auc_roc')
plt.title('Зависимость auc_roc модели  глубины')
plt.show()
```

Как видно из графика,что переобучение началось уже при глубине равной 9,наилучший показатель F1-меры получился при глубине 9 и равен F1-мера:0.5726375176304653,а лучшее значение auc_roc при глубине 6 .Посторим еще confusion_matrix

```python
# заполним предсказаниями 
exited_pred = best_tree.predict(features_valid)
recall = recall_score(target_valid, exited_pred)
#построим матрицу предсказаний 
conf_mx = confusion_matrix(target_valid, exited_pred)
ax = sns.heatmap(conf_mx, annot=True, fmt='4.0f', cmap='crest')
plt.title('confusion matrix', size=12)
ax.set(xlabel="Клиент уйдет:Да-1,Нет-0 \n \n Recall={r:.2f} ".format(r=recall),
       ylabel="Истинные значения")
```

### Случайный лес 

```python
#для лучшей 
best_forest_model = None
best_forest_est = 0
best_forest_depth = 0
best_f1_rf = 0
best_auc_roc_rf = 0

# для графиков списки 
auc_roc_forest = []
f1_rf_valid_forest_depth = []
forest_n_est = []
for est in range(20, 150,5): 
    forest_n_est.append(est)
    best_result_in_depth =[]
    auc_roc_forest_in_depth =[]
    for depth in range (7,20,2):
      # модель
        model_forest = RandomForestClassifier(
            random_state=12345, 
            n_estimators=est,
             max_depth=depth,
             n_jobs = -1)
        model_forest.fit(features_train, target_train)
        # метрика 
        pred_valid_forest = pd.Series(
            model_forest.predict(features_valid))
        f1_valid = f1_score(
            target_valid, pred_valid_forest) 
        
        #AUC-ROC
        probab_valid = model_forest.predict_proba(features_valid)
        prob_one = probab_valid[:, 1]
        auc_roc_valid = roc_auc_score(target_valid, prob_one)
        # лучшая модель 
        if f1_valid > best_f1_rf:
            best_forest_model = model_forest 
            best_forest_est = est 
            best_forest_depth = depth 
            best_f1_rf = f1_valid
            best_auc_roc_rf = auc_roc_valid

        auc_roc_forest_in_depth.append(auc_roc_valid)    
        best_result_in_depth.append(f1_valid)

    auc_roc_forest.append(max(auc_roc_forest_in_depth))
    f1_rf_valid_forest_depth.append(max(best_result_in_depth))

print('Лучшая модель Случайный лес:',best_forest_model)
print('Количество деревьев:', best_forest_est)
print('Глубина дерева лучшей модели:', best_forest_depth)
print('F1-мера лучшей модели:', best_f1_rf)
print('AUC-ROC лучшей модели:', best_auc_roc_rf)
```

Графики мои графики)

```python
x = forest_n_est
y = f1_rf_valid_forest_depth
plt.figure(figsize=(12, 4))
plt.grid(True)
plt.xticks(x)
plt.plot(x, y,'-bo')
plt.xlabel('n_estimators')
plt.ylabel('F1-мера')
plt.title('Зависимость F1-мера модели  от количества деревьев')
plt.show()
```

Как видно что наилучшее значение F1-меры достигается при количестве деревьев 35 и глубине 15

### Upsampling

```python
# как и в тренажере размножаем класс митозом 
def upsample(features, target, repeat = 4):
    # делим по классам
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    
    # повторяем и соединяем  
    
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    # мешаем 
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled
```

###  Downsampling

аналогично предыдущему,только для выборки выше и в обратную сторону 

```python
def downsample(features, target, fraction = 0.25):
    #делим 
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    #уменьшаем и соеденяем 
    
    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
        
    # перемешиваем
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
        
    return features_downsampled, target_downsampled



```

```python
features_train_down, target_train_down = downsample(
    features_train, target_train)
```

```python
y = target_train_down.value_counts()
fig1, ax1 =  plt.subplots()

ax1.pie(y,
        autopct=make_autopct(y),
        shadow=True,
        labels=['Не уйдет:',
                'Уйдет:'],
        colors = ['#66b3ff',
                  '#99ff99'],
        startangle=12.,
       explode=(0.1,0))
ax1.axis('equal')  
plt.title('Баланс классов downsample:',fontsize=12)
plt.show()
```

```python
#поиск лучшего 
best_tree_up = None
best_depth_tree_up = 0
best_f1_tree_up = 0
best_auc_roc_tree_up = 0
# для графиков 
depth_tree = []
f1_train_tree = []
f1_valid_tree_up = []
auc_roc_tree_list_up = []
auc_roc_tree_list_features =[]
#создаем 
for depth in range(1, 21, 1):
    model = DecisionTreeClassifier(random_state=12345, max_depth=depth
                                         )
    model.fit(features_train_upsampled, target_train_upsampled) 

    # получаем  f1-меру для обучающей и валидационной 
    pred_train_tree = pd.Series( model.predict(features_train_upsampled))
    f1_train = f1_score(target_train_upsampled, pred_train_tree)  

    pred_valid_tree = pd.Series(model.predict(features_valid))
    f1_valid = f1_score(target_valid, pred_valid_tree)   

    # добовляем 
    depth_tree.append(depth)
    f1_train_tree.append(f1_train)
    f1_valid_tree_up.append(f1_valid)
    
    #ищем auc_roc
    prob_valid = model.predict_proba(features_valid)
    prob_one_valid = prob_valid[:, 1]
    auc_roc_tree = roc_auc_score(target_valid,prob_one_valid)
    auc_roc_tree_list_up.append(auc_roc_tree)

    prob_features_train = model.predict_proba(features_train_upsampled)
    prob_one= prob_features_train[:, 1]
    auc_roc_tree_f = roc_auc_score(target_train_upsampled,prob_one)
    auc_roc_tree_list_features.append(auc_roc_tree_f)
    #ищем лучшую
    if f1_valid > best_f1_tree_up:
      best_tree_up = model
      best_depth_tree_up = depth
      best_f1_tree_up = f1_valid
      best_auc_roc_tree_up = auc_roc_tree
print('лучшая модель "дерево решений" upsampled:',best_tree_up)
print('Глубина дерева "дерево решений" upsampled:',best_depth_tree_up)
print('F1-мера "дерево решений" upsampled:',best_f1_tree_up)
print('AUC-ROC лучшей модели "дерево решений" upsampled:',best_auc_roc_tree_up)     
```
